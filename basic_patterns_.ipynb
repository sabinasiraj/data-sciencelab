```python
n = 5
for i in range(n):
    for j in range(n):
        print("*", end=" ")
    print()
```

```python
n = 5
for i in range(1, n + 1):
    print("* " * i)
```

```python
n = 5
num = 1
for i in range(1, n + 1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

```python
n = 5
for i in range(1, n + 1):
    for j in range(i):
        print(i, end=" ")
    print()
```

```python
n = 5
for i in range(n, 0, -1):
    print("* " * i)
```

```python
n = 5
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```
